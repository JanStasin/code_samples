```python
import os
import pandas as pd
from PIL import Image

image_folder = r'C:\Users\MSI\Desktop\DataScientest\CourseDatascientest\Files\images\images\image_train'


# Ensure imageid column is string
df1['imageid'] = df1['imageid'].astype(str)

# (Optional) Append '.jpg' if missing
df1['imageid'] = df1['imageid'].apply(lambda x: x if x.endswith('.jpg') else x + '.jpg')

# Function to load image
def load_image(image_id):
    image_path = os.path.join(image_folder, image_id)
    if os.path.exists(image_path):
        try:
            return Image.open(image_path)
        except Exception as e:
            print(f"Failed to open {image_id}: {e}")
            return None
    else:
        print(f"File not found: {image_path}")
        return None

# Apply the function
df1['image'] = df1['imageid'].apply(load_image)

# Show results
print(df1[['imageid', 'image']].head())

```

```python
#Chi-Square Test: categorical variables are dependent or not 


import pandas as pd
from scipy.stats import chi2_contingency
p= 0.05
table = pd.crosstab(df1['designation'], df1['imageid'].isnull())
chi2, p, dof, expected = chi2_contingency(table)
print(f"Chi-squared test p-value: {p:.4f}")

```
